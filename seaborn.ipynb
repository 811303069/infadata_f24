```python
sns.relplot(
    data = titanic,
    x = 'age',y = 'fare'
)
```

```python
titanic['Sex'] = titanic['sex'].map({"male": "Male", "female": "Female"})

sns.relplot(
    data = titanic,
    x = 'age',y = 'fare',
    alpha = 0.5, hue='Sex', col= 'survived',
    palette= {"Male" : "blue", "Female" : "magenta"}
)
```

```python
sns.relplot(
    data = titanic,
    x = 'age',y = 'fare',
    alpha = 0.5, hue='sex', col= 'survived',
    palette= {"male" : "blue", "female" : "magenta"},
    size= "pclass"
)
```

```python
sns.relplot(
    data = titanic,
    x = 'age',y = 'fare',
    alpha = 0.5, hue='sex', col= 'survived',
    palette= {"male" : "#e39381", "female" : "#555150"},
    size= "pclass"
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
smh = pd.read_csv("C:/Users/khush/Downloads/archive (4)/Student Mental health.csv")
smh.head()
```

```python
smh.groupby("Choose your gender")[['Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?']].count()
```

```python
clm_Plot = smh.columns[-4:-1]

category_order = ['Yes', 'No']
```

```python
for column in clm_Plot: 
    counts = smh[column].value_counts()
    counts = counts.reindex(category_order, fill_value = 0)
    print(counts)
    print()
```

```python
sns.countplot(smh, x='Do you have Depression?', stat ='percent', hue= 'Did you seek any specialist for a treatment?')
```
