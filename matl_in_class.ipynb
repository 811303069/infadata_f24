```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/khush/Downloads/cces_example_data.csv")
df.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].set_title("Age Histogram")
ax[1].set_title("Political Identity")

plt.show()
```

```python
evaluation_counts = df['Evaluation'].value_counts()
evaluation_counts.plot(kind='bar', figsize=(8, 6))

plt.title('Frequencies of Responses in Evaluation')
plt.xlabel('Evaluation Categories')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['ConMediaUse'], df['ImmiPhobia'], alpha=0.5)
plt.title('ConMediaUse vs ImmiPhobia')
plt.xlabel('ConMediaUse')
plt.ylabel('ImmiPhobia')

plt.show()
```
